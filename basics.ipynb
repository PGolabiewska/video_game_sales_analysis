# Video Game Sales - Basic Data

```python
import numpy as np
import pandas as pd
import matplotlib as plt
pd.set_option("display.width", 1000)

df = pd.read_csv("data\\games_data.csv")
print(df.drop(columns=["img"]).to_string)
```

Top 10 games on PC

```python
best = df.loc[df["console"] == "PC", ["title", "console", "genre", "publisher", "critic_score"]]
best = best.sort_values("critic_score", ascending=False).reset_index(drop=True)
best.index = range(1, len(best) + 1)

print(best.head(10))
```

10 worst games on PC

```python
worst = df.loc[df["console"] == "PC", ["title", "console", "genre", "publisher", "critic_score"]]
worst = worst.sort_values("critic_score", ascending=True).reset_index(drop=True)
worst.index = range(1, len(worst) + 1)

print(worst.head(10))
```
