Analyzing the results allows for addressing key business questions: 
- Identifying bestsellers—titles achieving the highest global sales
- Temporal trends—determining which years saw the largest market share and whether the industry is volatile
- Hardware specialization—whether specific consoles are linked to particular game genres
- Regional preferences—games that are hits in one region but fail in others.


```python
import numpy as np
import pandas as pd
pd.set_option("display.width", 1000)

df = pd.read_csv("data\\games_data.csv")
df = df.drop(columns=["img"])

print(df)
```

Dataset info

```python
df.info()
df.describe()
```

Top 10 games on PC

```python
best = df.loc[df["console"] == "PC", ["title", "console", "genre", "publisher", "critic_score"]]
best = best.sort_values("critic_score", ascending=False).reset_index(drop=True)
best.index = range(1, len(best) + 1)

print(best.head(10))
```

10 worst games on PC

```python
worst = df.loc[df["console"] == "PC", ["title", "console", "genre", "publisher", "critic_score"]]
worst = worst.sort_values("critic_score", ascending=True).reset_index(drop=True)
worst.index = range(1, len(worst) + 1)

print(worst.head(10))
```

Numbers of data

```python
df_title = df.drop_duplicates(subset=["title"])
df_console = df.drop_duplicates(subset=["console"])
df_genre = df.drop_duplicates(subset=["genre"])

print("Number of games: ",len(df_title))
print("Number of consoles: ",len(df_console))
print("Number of genres: ",len(df_genre))
```
